```python
#DESCRIPTIVE STATISTICS
import pandas as pd
df = pd.read_csv('retail_sales_dataset.csv')
mean_sales = df['Total Amount'].mean()
median_sales = df['Total Amount'].median()
mode_sales = df['Total Amount'].mode()
std_dev_sales = df['Total Amount'].std()
print("Mean:", mean_sales)
print("Median:", median_sales)
print("Mode:", mode_sales)
print("Standard Deviation:", std_dev_sales)
```

```python
#TIME SERIES ANALYSIS
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

df = pd.read_csv('retail_sales_dataset.csv')

df['Date'] = pd.to_datetime(df['Date'])

sales_over_time = df.groupby('Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(sales_over_time['Date'], sales_over_time['Total Amount'], color='black', linestyle='-')
plt.title('Total Sales Trends over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')

date_formatter = DateFormatter('%b %Y')
plt.gca().xaxis.set_major_formatter(date_formatter)
plt.xticks(rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#CUSTOMER AND PRODUCT ANALYSIS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('retail_sales_dataset.csv')

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Gender', y='Total Amount', estimator=sum)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(14, 6))
sns.histplot(data=df, x='Age', y='Total Amount',color='red')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Product Category', y='Total Amount',color='purple', estimator=sum)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation between product categories and customer demographics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('retail_sales_dataset.csv')

heatmap_data = df.groupby(['Product Category', 'Age']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 4))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Sales Distribution across Product Categories and Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Product Category')
plt.show()
```
